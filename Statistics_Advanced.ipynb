1. What is a random variable in probability theory?
  - A random variable is a function that assigns numerical values to the outcomes of a random experiment. It maps each possible outcome from a sample space to a real number, allowing us to quantify uncertain events mathematically.
2.  What are the types of random variables?
  - There are two main types:

  - 1. Discrete Random Variables: Take on countable values (e.g., number of heads in coin flips, dice rolls)
  - 2. Continuous Random Variables: Take on infinite values within an interval (e.g., height, weight, temperature)    
3. Explain the difference between discrete and continuous distributions.
  - Discrete distributions describe variables with countable outcomes and use probability mass functions (PMF). Examples include binomial and Poisson distributions.

  - Continuous distributions describe variables with infinite possible values in a range and use probability density functions (PDF). Examples include normal and exponential distributions.
4.  What is a binomial distribution, and how is it used in probability?
  - A binomial distribution models the number of successes in a fixed number of independent trials, each with the same probability of success. It requires: fixed number of trials (n), two possible outcomes, constant probability (p), and independent trials. Used in quality control, medical trials, and survey analysis.
5.  What is the standard normal distribution, and why is it important?
  - The standard normal distribution is a normal distribution with mean = 0 and standard deviation = 1. It's important because any normal distribution can be converted to it using z-scores, enabling standardized probability calculations and hypothesis testing.
6. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - The CLT states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, regardless of the population's distribution. It's critical because it justifies using normal distribution methods for inference, enables confidence intervals and hypothesis testing, and works with samples as small as 30.
7.  What is the significance of confidence intervals in statistical analysis?
  - Confidence intervals provide a range of plausible values for a population parameter with a specified confidence level (e.g., 95%). They quantify uncertainty in estimates, are more informative than point estimates, and help in decision-making and hypothesis testing.
8. What is the concept of expected value in a probability distribution?
  - The expected value (E[X]) is the long-run average value of a random variable over many repetitions. For discrete variables: E[X] = Σ(x × P(x)). For continuous variables: E[X] = ∫x × f(x)dx. It represents the center of the distribution.

          

```python
'''#9. Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution. '''
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=1000)


mean = np.mean(data)
std_dev = np.std(data, ddof=1)

print(f"Mean: {mean:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")

# Draw histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f'Mean = {mean:.2f}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of 1000 Random Numbers from Normal Distribution\n(μ=50, σ=5)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

10. Question : ( You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
 - daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
- 10.1 : Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
 - 10.2 : Write the Python code to compute the mean sales and its confidence interval. )
 - Answer: The CLT allows us to assume the sampling distribution of the mean is approximately normal. With the sample mean and standard error, we can construct a 95% confidence interval using: CI = mean ± (critical_value × standard_error).


```python
# 10.2 :
import numpy as np
from scipy import stats
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
n = len(daily_sales)
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)
standard_error = std_sales / np.sqrt(n)
confidence_level = 0.95
alpha = 1 - confidence_level
df = n - 1
t_critical = stats.t.ppf(1 - alpha/2, df)
margin_of_error = t_critical * standard_error
ci_lower = mean_sales - margin_of_error
ci_upper = mean_sales + margin_of_error
# Display results
print(f"Sample Size: {n}")
print(f"Mean Sales: ${mean_sales:.2f}")
print(f"Standard Deviation: ${std_sales:.2f}")
print(f"Standard Error: ${standard_error:.2f}")
print(f"95% Confidence Interval: (${ci_lower:.2f}, ${ci_upper:.2f})")
print(f"\nInterpretation: We are 95% confident that the true average daily sales")
print(f"lies between ${ci_lower:.2f} and ${ci_upper:.2f}")
```
