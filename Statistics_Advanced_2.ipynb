# Assignment Questions

1. What is hypothesis testing in statistics?
  - Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves formulating two competing hypotheses (null and alternative), collecting sample data, and using statistical tests to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis. The process helps researchers determine whether observed patterns in data are statistically significant or could have occurred by chance.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - The null hypothesis (H₀) is a statement of no effect, no difference, or no relationship between variables. It represents the status quo or default assumption that any observed difference is due to random chance.

  - The alternative hypothesis (H₁ or Hₐ) is a statement that contradicts the null hypothesis, suggesting there is a significant effect, difference, or relationship.

  - Key differences:

  - H₀ assumes no change or effect; H₁ assumes there is a change or effect
We test to reject H₀, not to prove it
Example: H₀: μ = 50 (population mean equals 50) vs. H₁: μ ≠ 50 (population mean does not equal 50)
3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
  - The significance level (α) is the probability threshold for rejecting the null hypothesis when it is actually true (Type I error). Commonly set at 0.05 (5%), 0.01 (1%), or 0.10 (10%).

  - Role in decision-making:

  - If the p-value ≤ α: Reject H₀ (result is statistically significant)
  - If the p-value > α: Fail to reject H₀ (insufficient evidence)
  - The significance level represents the researcher's tolerance for making a Type I error. A lower α (e.g., 0.01) requires stronger evidence to reject H₀, reducing false positives but potentially missing true effects. The choice of α should be made before conducting the test based on the context and consequences of errors.
4. What are Type I and Type II errors? Give examples of each.
  - Type I Error (α): Rejecting the null hypothesis when it is actually true (false positive).

  - Example: A medical test indicates a patient has a disease when they are actually healthy. The null hypothesis (patient is healthy) is incorrectly rejected.
  - Type II Error (β): Failing to reject the null hypothesis when it is actually false (false negative).

  - Example: A medical test indicates a patient is healthy when they actually have a disease. The null hypothesis (patient is healthy) is incorrectly retained.
Summary:

  - Type I: False alarm, concluding an effect exists when it doesn't
  - Type II: Missed detection, failing to identify a real effect
  - There's a trade-off: reducing one type of error often increases the other
5. What is the difference between a Z-test and a T-test? Explain when to use each.
  - -Z-test:

  - Uses the standard normal distribution (Z-distribution)
  - Assumes population standard deviation (σ) is known
  - Appropriate for large sample sizes (n ≥ 30)
  - Test statistic: Z = (x̄ - μ) / (σ/√n)
  - T-test:

  - Uses the Student's t-distribution
  - Uses sample standard deviation (s) when population σ is unknown
  - Appropriate for small sample sizes (n < 30)
  - Test statistic: t = (x̄ - μ) / (s/√n)
  - Has heavier tails than normal distribution, accounting for additional uncertainty
  - When to use:

  - Z-test: Large samples (n ≥ 30) OR known population standard deviation
  - T-test: Small samples (n < 30) AND unknown population standard deviation  
    



```python
'''# 6. Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram. '''
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)


n = 10
p = 0.5
size = 1000


binomial_data = np.random.binomial(n, p, size)

plt.figure(figsize=(10, 6))
plt.hist(binomial_data, bins=range(0, n+2), edgecolor='black', alpha=0.7, density=True)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xticks(range(0, n+1))
plt.grid(axis='y', alpha=0.3)
plt.show()


print(f"Mean: {np.mean(binomial_data):.2f}")
print(f"Standard Deviation: {np.std(binomial_data):.2f}")
print(f"Theoretical Mean: {n*p:.2f}")
print(f"Theoretical Std Dev: {np.sqrt(n*p*(1-p)):.2f}")
```

```python
'''# 7. Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
'''
import numpy as np
from scipy import stats


sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


population_mean = 50
alpha = 0.05


sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)
standard_error = sample_std / np.sqrt(n)


z_statistic = (sample_mean - population_mean) / standard_error


p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))


z_critical = stats.norm.ppf(1 - alpha/2)


print("=" * 50)
print("HYPOTHESIS TESTING USING Z-STATISTICS")
print("=" * 50)
print(f"Null Hypothesis (H0): μ = {population_mean}")
print(f"Alternative Hypothesis (H1): μ ≠ {population_mean}")
print(f"Significance Level (α): {alpha}")
print("\n" + "-" * 50)
print("SAMPLE STATISTICS:")
print("-" * 50)
print(f"Sample Size (n): {n}")
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Std Dev: {sample_std:.4f}")
print(f"Standard Error: {standard_error:.4f}")
print("\n" + "-" * 50)
print("TEST RESULTS:")
print("-" * 50)
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical Z-values: ±{z_critical:.4f}")
print("\n" + "-" * 50)
print("DECISION:")
print("-" * 50)
if p_value < alpha:
    print(f"P-value ({p_value:.4f}) < α ({alpha})")
    print("REJECT the null hypothesis")
    print(f"Conclusion: There is significant evidence that the population mean differs from {population_mean}")
else:
    print(f"P-value ({p_value:.4f}) ≥ α ({alpha})")
    print("FAIL TO REJECT the null hypothesis")
    print(f"Conclusion: There is insufficient evidence that the population mean differs from {population_mean}")
print("=" * 50)
```

```python
'''# 8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
'''
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


np.random.seed(42)


mu = 100
sigma = 15
n = 50


sample_data = np.random.normal(mu, sigma, n)


sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
standard_error = sample_std / np.sqrt(n)


confidence_level = 0.95
alpha = 1 - confidence_level
z_critical = stats.norm.ppf(1 - alpha/2)


margin_of_error = z_critical * standard_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error


print("=" * 60)
print("95% CONFIDENCE INTERVAL FOR POPULATION MEAN")
print("=" * 60)
print(f"Sample Size: {n}")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Std Dev: {sample_std:.2f}")
print(f"Standard Error: {standard_error:.2f}")
print(f"\n95% Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}]")
print(f"Margin of Error: ±{margin_of_error:.2f}")
print("=" * 60)


fig, axes = plt.subplots(1, 2, figsize=(14, 5))


axes[0].hist(sample_data, bins=15, edgecolor='black', alpha=0.7, color='skyblue')
axes[0].axvline(sample_mean, color='red', linestyle='--', linewidth=2, label=f'Sample Mean: {sample_mean:.2f}')
axes[0].axvline(mu, color='green', linestyle='--', linewidth=2, label=f'True Mean: {mu:.2f}')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Sample Data Distribution')
axes[0].legend()
axes[0].grid(axis='y', alpha=0.3)


axes[1].errorbar(1, sample_mean, yerr=margin_of_error, fmt='o', markersize=10,
                 capsize=10, capthick=2, color='blue', label='Sample Mean with 95% CI')
axes[1].axhline(mu, color='green', linestyle='--', linewidth=2, label=f'True Mean: {mu:.2f}')
axes[1].axhline(ci_lower, color='red', linestyle=':', alpha=0.5)
axes[1].axhline(ci_upper, color='red', linestyle=':', alpha=0.5)
axes[1].fill_between([0.5, 1.5], ci_lower, ci_upper, alpha=0.2, color='blue')
axes[1].set_xlim(0.5, 1.5)
axes[1].set_ylabel('Value')
axes[1].set_title('95% Confidence Interval')
axes[1].legend()
axes[1].grid(axis='y', alpha=0.3)
axes[1].set_xticks([])

plt.tight_layout()
plt.show()
```

```python
'''# 9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
'''
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)


data = np.random.normal(loc=75, scale=10, size=100)

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    z_scores = (data - mean) / std_dev
    return z_scores, mean, std_dev


z_scores, mean, std_dev = calculate_z_scores(data)


print("=" * 60)
print("Z-SCORE ANALYSIS")
print("=" * 60)
print(f"Original Data Statistics:")
print(f"  Mean: {mean:.2f}")
print(f"  Standard Deviation: {std_dev:.2f}")
print(f"  Min: {np.min(data):.2f}")
print(f"  Max: {np.max(data):.2f}")
print(f"\nZ-Score Statistics:")
print(f"  Mean: {np.mean(z_scores):.6f} (should be ~0)")
print(f"  Standard Deviation: {np.std(z_scores, ddof=1):.6f} (should be ~1)")
print(f"  Min Z-score: {np.min(z_scores):.2f}")
print(f"  Max Z-score: {np.max(z_scores):.2f}")
print("\n" + "-" * 60)
print("Z-SCORE INTERPRETATION:")
print("-" * 60)
print("Z-scores represent the number of standard deviations a data")
print("point is from the mean:")
print("  • Z = 0: Value equals the mean")
print("  • Z = 1: Value is 1 std dev above the mean")
print("  • Z = -1: Value is 1 std dev below the mean")
print("  • Z = 2: Value is 2 std devs above the mean")
print("  • |Z| > 3: Potential outlier (very rare)")
print("=" * 60)


fig, axes = plt.subplots(2, 2, figsize=(14, 10))


axes[0, 0].hist(data, bins=20, edgecolor='black', alpha=0.7, color='lightblue')
axes[0, 0].axvline(mean, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean:.2f}')
axes[0, 0].axvline(mean + std_dev, color='orange', linestyle='--', linewidth=1.5, label=f'+1 SD: {mean+std_dev:.2f}')
axes[0, 0].axvline(mean - std_dev, color='orange', linestyle='--', linewidth=1.5, label=f'-1 SD: {mean-std_dev:.2f}')
axes[0, 0].set_xlabel('Original Values')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].set_title('Original Data Distribution')
axes[0, 0].legend()
axes[0, 0].grid(axis='y', alpha=0.3)


axes[0, 1].hist(z_scores, bins=20, edgecolor='black', alpha=0.7, color='lightgreen')
axes[0, 1].axvline(0, color='red', linestyle='--', linewidth=2, label='Mean: 0')
axes[0, 1].axvline(1, color='orange', linestyle='--', linewidth=1.5, label='+1 SD')
axes[0, 1].axvline(-1, color='orange', linestyle='--', linewidth=1.5, label='-1 SD')
axes[0, 1].set_xlabel('Z-scores')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].set_title('Standardized Data (Z-scores)')
axes[0, 1].legend()
axes[0, 1].grid(axis='y', alpha=0.3)


axes[1, 0].scatter(data, z_scores, alpha=0.6, color='purple')
axes[1, 0].axhline(0, color='red', linestyle='--', linewidth=1)
axes[1, 0].axvline(mean, color='red', linestyle='--', linewidth=1)
axes[1, 0].set_xlabel('Original Values')
axes[1, 0].set_ylabel('Z-scores')
axes[1, 0].set_title('Original Values vs Z-scores')
axes[1, 0].grid(True, alpha=0.3)


box_data = [data, z_scores]
bp = axes[1, 1].boxplot(box_data, labels=['Original Data', 'Z-scores'], patch_artist=True)
bp['boxes'][0].set_facecolor('lightblue')
bp['boxes'][1].set_facecolor('lightgreen')
axes[1, 1].set_ylabel('Values')
axes[1, 1].set_title('Box Plot Comparison')
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()


print("\nSample Z-score Examples:")
print("-" * 60)
for i in range(5):
    print(f"Original Value: {data[i]:.2f} \u2192 Z-score: {z_scores[i]:.2f}")
    if z_scores[i] > 0:
        print(f"  \u2192 {abs(z_scores[i]):.2f} standard deviations ABOVE the mean")
    else:
        print(f"  \u2192 {abs(z_scores[i]):.2f} standard deviations BELOW the mean")
```
