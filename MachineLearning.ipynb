1. What is a parameter?
  - A parameter is a variable in a model that is learned from the training data. In machine learning, parameters are the internal coefficients or weights that the model adjusts during training to minimize error (e.g., weights in linear regression, neural network weights).
2. What is correlation? What does negative correlation mean?
  - Correlation is a statistical measure that describes the strength and direction of the relationship between two variables. It ranges from -1 to +1, where values closer to ±1 indicate stronger relationships.
  - Negative correlation means that as one variable increases, the other variable tends to decrease. For example, as the price of a product increases, demand typically decreases.

4. How does loss value help in determining whether the model is good or not?

13. What does negative correlation mean?
  - Negative correlation (correlation coefficient between -1 and 0) indicates an inverse relationship where as one variable increases, the other variable tends to decrease proportionally. Example: As study time increases, the number of errors typically decreases.
    

```python
#14. How can you find correlation between variables in Python?
import pandas as pd
import numpy as np

# Example: Create a sample DataFrame 'df' to demonstrate correlation calculation
data = {
    'var1': [1, 2, 3, 4, 5],
    'var2': [5, 4, 3, 2, 1],
    'var3': [1, 2, 1, 2, 1]
}
df = pd.DataFrame(data)

correlation_matrix = df.corr()
print("Correlation Matrix:\n", correlation_matrix)

correlation_var1_var2 = df['var1'].corr(df['var2'])
print(f"Correlation between var1 and var2 (using .corr()): {correlation_var1_var2}")

correlation_np = np.corrcoef(df['var1'], df['var2'])[0, 1]
print(f"Correlation between var1 and var2 (using numpy.corrcoef): {correlation_np}")

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Variables')
plt.show()
```

15. What is causation? Explain difference between correlation and causation with an example.
  - Causation means one variable directly causes changes in another variable - there is a cause-and-effect relationship.

  - Difference:

  - Correlation: Statistical relationship between variables (they move together)
  - Causation: One variable directly influences or causes the other
    - Example: Ice cream sales and drowning incidents are highly correlated (both increase in summer), but ice cream sales don't cause drowning. The real cause is warm weather, which is a confounding variable that affects both.

21. What is feature scaling? How does it help in Machine Learning?
  - Feature scaling is the process of transforming features to a similar scale or range.

  - How it helps:

     - Prevents dominance: Features with larger ranges don't dominate smaller ones
     - Faster convergence: Gradient descent converges more quickly
     - Improves accuracy: Distance-based algorithms (KNN, SVM, K-means) perform better
     - Regularization: Makes regularization penalties fair across features
     - Neural networks: Helps with stable and faster training
