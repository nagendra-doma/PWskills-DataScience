1.What is the difference between descriptive statistics and inferential
statistics? Explain with examples.
  - Descriptive statistics aim to summarize and describe the main features of a dataset that has already been collected. They deal with facts about the known data (either a whole population or a sample).
  - Example : Imagine a high school principal records the final exam scores of all 100 students in a specific math course.

  - Calculation: The principal finds the average score (mean) is 78%.

  - Presentation: They create a bar chart showing the number of students who scored in each grade band (A, B, C, etc.).

  - Conclusion: The principal is only stating a fact about that specific group of 100 students in that course. They are not making any claims about the scores of students in other courses or schools.


  - Inferential statistics aim to make predictions or generalizations about a larger population based on a smaller representative sample. They involve drawing conclusions that go beyond the data immediately available.
  - Example : Imagine a national education board wants to know the average math score of all high school students across the country (millions of students).

  - Sampling: It's impossible to test every student, so the board selects a random sample of 1,000 students from various schools.

  - Calculation: They find the average score of this sample is 75%.

  - Inference: Using inferential statistics (like a confidence interval), they conclude, "We are 95% confident that the true average math score for all high school students in the country is between 73% and 77%."

  - Conclusion: They are using the small sample data to make a generalization (an inference) about the entire national student population, with an acknowledged margin of error
  2. What is sampling in statistics? Explain the differences between random
and stratified sampling.
  - Sampling in statistics is the process of selecting a representative subset (a sample) from a larger group (a population) to gather data and make inferences about the entire population. It's often impractical or impossible to study every member of a population, so sampling allows researchers to collect data more efficiently and at a lower cost.
  - Stratified sampling involves dividing the population into subgroups (strata) based on shared characteristics (e.g., age, gender, income level, geographic location) and then randomly selecting a sample from each stratum. The proportion of the sample taken from each stratum is usually proportional to the stratum's size in the population.
  3. Define mean, median, and mode. Explain why these measures of central
tendency are important.
  - Measures of central tendency are single values that describe the center point or typical value of a dataset. They help to summarize and understand the distribution of data. The most common measures are mean, median, and mode.
  - Mean:

  - Definition: The mean, also known as the average, is calculated by summing all the values in a dataset and dividing by the number of values.
  - Calculation: Sum of all values / Number of values
  - Importance: The mean is widely used because it takes into account every value in the dataset. It's particularly useful for data that is symmetrically distributed without extreme outliers.
  - Median:

  - Definition: The median is the middle value in a dataset when the data is arranged in ascending or descending order. If there is an even number of values, the median is the average of the two middle values.
  - Calculation: Arrange data in order and find the middle value(s).
  - Importance: The median is a robust measure of central tendency because it is not affected by extreme outliers. It's a good choice for skewed data or when outliers are present.
  - Mode:
  - Definition: The mode is the value that appears most frequently in a dataset. A dataset can have one mode (unimodal), more than one mode (multimodal), or no mode if all values appear with the same frequency.
  - Calculation: Identify the value(s) with the highest frequency.
  - Importance: The mode is useful for identifying the most common category or value in a dataset. It is the only measure of central tendency that can be used with categorical data.
  4. Explain skewness and kurtosis. What does a positive skew imply about the
data?
  - Skewness is a measure of the asymmetry, or lack of symmetry, in a data distribution.

  - A distribution is perfectly symmetric (like the normal distribution or "bell curve") if it looks the same on both the left and right sides of the center point; its skewness value is zero.

  - It tells you which side of the distribution has a longer "tail."
  - kurtosis :     
  - Kurtosis is a measure of the "tailedness" of a distribution relative to a normal distribution. In simpler terms, it indicates the degree to which data points cluster around the center and how thick the tails are (i.e., the presence of outliers).
  - A positive skew (or right-skew) implies the following about the data distribution :

  - Tail Direction: The distribution has a longer, stretched-out tail on the right side. This means there are a few unusually large values (outliers) that pull the distribution's shape in that direction.


  - Data Concentration: The majority of the data is clustered towards the lower (left) end of the distribution.

  - Cen tral Tendency Relationship: The mean, median, and mode are generally in this order:

  - Mode < Median < Mean
  - The Mode (most frequent value) is the smallest, at the highest point of the peak.

  - The Median (middle value) is next.

  - The Mean (average) is the largest because the few extreme high values in the right tail pull the average toward them.

  - A classic example is household income—most people earn a modest amount (clustered on the left), but a few individuals earn extremely high amounts (the long right tail), causing the mean income to be higher than the median income.
  

```python
'''# 5. Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]'''
import statistics
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
mean_value = statistics.mean(numbers)
median_value = statistics.median(numbers)
mode_value = statistics.mode(numbers)
print(f"The given list of numbers is: {numbers}")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value}")
```

```python
'''# 6. Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]'''
import numpy as np
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]
x = np.array(list_x)
y = np.array(list_y)
covariance = np.cov(x, y)[0, 1]
correlation_coefficient = np.corrcoef(x, y)[0, 1]

print(f"Dataset X: {list_x}")
print(f"Dataset Y: {list_y}")
print(f"Covariance (Cov(X, Y)): {covariance}")
print(f"Correlation Coefficient (r): {correlation_coefficient}")
```

```python
'''# 7. Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]'''
import matplotlib.pyplot as plt
import numpy as np
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
np_data = np.array(data)
plt.figure(figsize=(6, 4))
plt.boxplot(np_data, vert=False, patch_artist=True, meanline=True, showmeans=True)
plt.title('Boxplot of Numeric Data')
plt.yticks([1], ['Data'])
plt.xlabel('Values')
plt.grid(axis='x', linestyle='--', alpha=0.6)
boxplot_filename = 'data_boxplot.png'
plt.savefig(boxplot_filename)
plt.close()
Q1 = np.percentile(np_data, 25)
Q3 = np.percentile(np_data, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = np_data[(np_data < lower_bound) | (np_data > upper_bound)]
print(f"Data: {data}")
print(f"Q1 (25th Percentile): {Q1}")
print(f"Q3 (75th Percentile): {Q3}")
print(f"IQR (Q3 - Q1): {IQR}")
print(f"Lower Bound for Outliers (Q1 - 1.5*IQR): {lower_bound}")
print(f"Upper Bound for Outliers (Q3 + 1.5*IQR): {upper_bound}")
print(f"Identified Outliers: {outliers.tolist()}")
print(f"Boxplot saved as: {boxplot_filename}")
```

8. You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
 Explain how you would use covariance and correlation to explore this
relationship?
  - 1. Covariance: Direction of the Relationship
  - Purpose: Covariance measures how two variables—Advertising Spend (X) and Daily Sales (Y)—change together. It tells us the direction of the relationship.
  - Positive Covariance (e.g., +1000): This means that as advertising spend increases, daily sales tend to increase as well.
  - Negative Covariance (e.g., -1000): This means that as advertising spend increases, daily sales tend to decrease.
  - Near Zero Covariance: This suggests little to no linear relationship between the two variables.
  - Limitation: Covariance's magnitude depends on the units of the data (dollars squared, in this case), making it hard to compare with other datasets or interpret the strength of the relationship.

   - 2. Correlation: Strength and Direction
  - Purpose: Correlation (specifically, Pearson's r) standardizes the covariance, creating a value between −1.0 and +1.0. This value indicates both the strength and the direction of the linear relationship.
  - r=+1.0: Perfect positive linear relationship (Sales increase directly with Spend).
  - r close to +1.0 (e.g., +0.95): Very strong positive linear relationship (Marketing should continue investing as sales strongly track the spend).
  - r close to −1.0 (e.g., −0.85): Strong negative linear relationship (This would be highly unusual and a cause for investigation!).
  - r close to 0: Weak or no linear relationship (Marketing spend may not be the primary driver of sales, or the relationship isn't linear).
  - Conclusion: For a data analyst, the correlation coefficient is more informative because its standardized value directly indicates how reliable the relationship is for predictive modeling and business decision-making.
  

```python
'''# 8. You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
●
Explain how you would use covariance and correlation to explore this
relationship.
●
Write Python code to compute the correlation between the two lists:
advertising_
spend = [200, 250, 300, 400, 500]
daily_
sales = [2200, 2450, 2750, 3200, 4000] '''
import numpy as np
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]
X = np.array(advertising_spend)
Y = np.array(daily_sales)
covariance = np.cov(X, Y)[0, 1]
correlation_coefficient = np.corrcoef(X, Y)[0, 1]
print(f"Advertising Spend (X): {advertising_spend}")
print(f"Daily Sales (Y): {daily_sales}")
print("-" * 50)
print(f"Computed Covariance (Cov(X, Y)): {covariance:.2f}")
print(f"Computed Correlation Coefficient (r): {correlation_coefficient:.6f}")
```

9. Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
Write Python code to create a histogram using Matplotlib .
For the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
  - Central Tendency	Mean	The average score. Highly useful for comparison, but sensitive to extreme scores.
  - Median	The middle score (50th percentile). Less sensitive to outliers, providing a more robust central value.
  - Dispersion	Standard Deviation (σ)	Measures the average amount of variability in the scores. A low σ means scores are clustered tightly around the mean; a high σ means scores are spread out.
  - Range	The difference between the highest and lowest scores.
  - Shape	Histogram :	The best way to visualize the overall distribution. It shows if the data is symmetric, skewed (e.g., heavily skewed toward the 10, which is common in satisfaction data), or multimodal.
  - Box Plot	Quickly shows the Median (center line), the Interquartile Range (box width/height), and potential outliers, offering a compact view of spread and center.									!

```python
'''# 9. Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
●
●
Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
Write Python code to create a histogram using Matplotlib
For the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]'''
import matplotlib.pyplot as plt
import numpy as np
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
mean_score = np.mean(survey_scores)
median_score = np.median(survey_scores)
std_dev = np.std(survey_scores)
print(f"Mean Score: {mean_score:.2f}")
print(f"Median Score: {median_score}")
print(f"Standard Deviation: {std_dev:.2f}")
plt.figure(figsize=(10, 6))
bins = np.arange(3.5, 10.6, 1)
plt.hist(survey_scores, bins=bins, edgecolor='black', alpha=0.7)
plt.xticks(np.arange(4, 11, 1))
plt.title('Distribution of Customer Satisfaction Survey Scores (Scale 1-10)', fontsize=14)
plt.xlabel('Satisfaction Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.5)
histogram_filename = 'satisfaction_scores_histogram.png'
plt.savefig(histogram_filename)
print(f"Histogram saved as {histogram_filename}")
```
